#Processing OFFSPEC Runs

`OFFSPEC` does not currently have a correct parameter file, so processing `OFFSPEC` data using `ReflectometryReductionOneAuto` requires many properties to be set.

Specifically:

```
DetectorComponentName="HeDetector",
SampleComponentName="sampleholder",
MonitorBackgroundWavelengthMin=15,
MonitorBackgroundWavelengthMax=20,
MonitorIntegrationWavelengthMin=2,
MonitorIntegrationWavelengthMax=14,
I0MonitorIndex=1,
ProcessingInstructions="3"
```


````python
#Set up the environment
import sys
import os
sys.path.append(os.environ['MANTIDPATH'])
from mantid.simpleapi import *
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#Load our input workspaces
offspec75 = Load("OFFSPEC00027575.nxs") #th=0.35
offspec76 = Load("OFFSPEC00027576.nxs") #th=1.00
offspec78 = Load("OFFSPEC00027578.nxs") #th=1.70
offspec85 = Load("OFFSPEC00027585.nxs") #transmission run

#Process using ReflectometryReductionOneAuto
ivq_75, _, _ = ReflectometryReductionOneAuto(offspec75, ThetaIn = 0.35, FirstTransmissionRun=offspec85, DetectorComponentName="HeDetector", SampleComponentName="sampleholder", MonitorBackgroundWavelengthMin=15, MonitorBackgroundWavelengthMax=20, MonitorIntegrationWavelengthMin=2, MonitorIntegrationWavelengthMax=14, I0MonitorIndex=1, ProcessingInstructions="3")
ivq_76, _, _ = ReflectometryReductionOneAuto(offspec76, ThetaIn = 1.00, FirstTransmissionRun=offspec85, DetectorComponentName="HeDetector", SampleComponentName="sampleholder", MonitorBackgroundWavelengthMin=15, MonitorBackgroundWavelengthMax=20, MonitorIntegrationWavelengthMin=2, MonitorIntegrationWavelengthMax=14, I0MonitorIndex=1, ProcessingInstructions="3")
ivq_78, _, _ = ReflectometryReductionOneAuto(offspec78, ThetaIn = 1.70, FirstTransmissionRun=offspec85, DetectorComponentName="HeDetector", SampleComponentName="sampleholder", MonitorBackgroundWavelengthMin=15, MonitorBackgroundWavelengthMax=20, MonitorIntegrationWavelengthMin=2, MonitorIntegrationWavelengthMax=14, I0MonitorIndex=1, ProcessingInstructions="3")

#Stitch the IvsQ workspaces together
ivq_75_76, _ = Stitch1D(ivq_75, ivq_76, Params="1e-3")
ivq_75_76_78, _ = Stitch1D(ivq_75_76, ivq_78, Params="0,1e-3,0.08")
````

````python
#Helper function to plot with
def plot_ws(axes, data, label):
    point_data = ConvertToPointData(data)
    axes.errorbar(point_data[0].readX(0), point_data[0].readY(0), yerr=point_data[0].readE(0), label=label)

#Use a figure, so we can specify the dimensions of the plot
fig = plt.figure(figsize=(14,5))
axes = fig.add_axes([0,0,1,1])
plot_ws(axes, ivq_75_76_78, "Stitched")
plot_ws(axes, ivq_75, "IvsQ 75")
plot_ws(axes, ivq_76, "IvsQ 76")
plot_ws(axes, ivq_78, "IvsQ 78")

axes.legend()
axes.grid()
axes.set_xlabel("Momentum transfer")
axes.set_ylabel("Counts")
axes.set_yscale("log")
````
